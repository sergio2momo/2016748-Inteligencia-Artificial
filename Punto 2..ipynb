Escriba una clase iterable que tenga la siguiente funcionalidad:
1. Al instanciar un objeto, lee el archivo de datos y lo coloca en un dataframe interno en la clase.
2. Cada vez que se le pidan datos, debe regresar los dos siguientes datos.
3. Imprima "error" si los datos se han agotado.

```python
#Importar librerÃ­as:

import pandas as pd
```

```python
#Construir la clase:

class ClassOne:
    def __init__(self, file_path):
        self.data = pd.read_csv(file_path)
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.data):
            raise StopIteration("error")
        else:
            row1, row2 = self.data.iloc[self.index], self.data.iloc[self.index+1]
            self.index += 2
            return row1, row2
```

```python
#Probar la clase:

iris_iterator = ClassOne("iris.csv")
for _ in range(82):
    try:
        data = next(iris_iterator)
        print(data)
    except StopIteration as e:
        print(e)
        break
```
